```python
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns 
```

### Bar Chart

```python
for column in categorical_columns:
    plt.figure(figsize=(6, 4))
    value_counts = df[column].value_counts()
    value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Bar Chart for {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('Frequency')
    plt.xticks(rotation=0)
    plt.show()
```

### Standardization of Numerical Variable

#### Z-score normalization is a crucial stem in data preprocessing.known as standard normalization, transforms the data into a standard normal distribution where the mean is 0 and the standard deviation is 1

```python
z_normalization=(df[numerical_columns]-df[numerical_columns].mean())/ df[numerical_columns].std()
```

### Conversion of Categorical data into Dummy Variables

#### Many machine learning algorithms cannot handle categorical data directly; they require numerical input. One-hot encoding is a method to convert categorical variables into a binary matrix (0s and 1s) where each category is represented as a separate column.

```python
one_hot_encoded=pd.get_dummies(df, columns=categorical_columns, prefix=categorical_columns, drop_first=False)
```
